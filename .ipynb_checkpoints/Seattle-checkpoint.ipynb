### Update Data Types

```python
df_reviews['date'] = pd.to_datetime(df_reviews['date'])
```

```python
df_listings['last_scraped'] = pd.to_datetime(df_listings['last_scraped'])
```

```python
df_calendar['date'] = pd.to_datetime(df_calendar['date'])
df_calendar['month'] = pd.DatetimeIndex(df_calendar['date']).month
df_calendar
```

```python
def update_true_false(t_or_f):
    if t_or_f == 't':
        return_value = True
    elif t_or_f == 'f':
        return_value = False
    else:
        return_value = t_or_f
        
    return return_value

def clean_price_field(price):
    #price = price.str.replace(np.nan, '0')
    if ( price == None ):
        price = 0
    
    if isinstance(price, float):
        if np.isnan(price) == True:
            price = 0
        
    if isinstance(price, str):
        price = price.replace(',', '')
        price = price.replace('$', '')
        
    price = float(price)
    return price
```

```python
#Update 'Available', 'Instant Bookable', 'Require Guest Profile' field to True / False
df_calendar['available'] = df_calendar.apply(lambda x: update_true_false(x['available']),axis=1)
df_listings['instant_bookable'] = df_listings.apply(lambda x: update_true_false(x['instant_bookable']),axis=1)
df_listings['require_guest_profile_picture'] = df_listings.apply(lambda x: update_true_false(x['require_guest_profile_picture']),axis=1)
```

```python
#Update Calendar 'Price' to float value. 
df_calendar['price'] = df_calendar.apply(lambda x: clean_price_field(x['price']),axis=1)
```

```python
#Add Deciles & Quartiles to Price Column
df_calendar['price_quartile'] = pd.qcut(df_calendar['price'].where(df_calendar['price'] > 0), q=4, precision=2)
df_calendar['price_deciles'] = pd.qcut(df_calendar['price'].where(df_calendar['price'] > 0), q=10, precision=2)
```

```python
#Update Listing 'Cleaning Fee' to Float Value
df_listings['cleaning_fee'] = df_listings.apply(lambda x: clean_price_field(x['cleaning_fee']),axis=1)
```

```python
#Create Dataframe Containing Only 'Available' == True Records
df_calendar_all_available = df_calendar[df_calendar['available'] == True]
```

```python
df_calendar_all_available.sample(15)
#df_calendar.sample(15)
```

```python
df_listings_small = df_listings[['id', 'host_id', 'name', 'neighbourhood_group_cleansed', 'cleaning_fee', 'instant_bookable']]
df_listings_small.head(20)
```

```python
#df_listings_small_calendar = pd.merge(df_calendar, df_listings_small, on=['listing_id', 'id'])
df_listings_small_calendar = pd.merge(df_calendar, df_listings_small, left_on='listing_id', right_on='id')
df_listings_small_calendar.drop(['id'], inplace=True, axis=1)
df_listings_small_calendar = df_listings_small_calendar.sort_values(by=['listing_id', 'date'])
df_listings_small_calendar.dtypes
```

```python
df_listings_small_calendar.sample(10)
```

```python
#Total Price for All Available Rooms
total_price_all_available_rooms = round(df_calendar_all_available['price'].sum(), 2)
print('Total Price for All Available Rooms: {:15,.2f}'.format(total_price_all_available_rooms))
```

```python
#Average # of Listings Per Night, By Neighborhood
df_listings_small_calendar[df_listings_small_calendar['available'] == True] \
    .groupby('neighbourhood_group_cleansed')['date'] \
    .count() \
    .sort_values(ascending=False) / 365
```

```python
#What is the average nightly price by neighborhood?
df_avg_nightly_by_neighborhood = df_listings_small_calendar[df_listings_small_calendar['available'] == True] \
    .groupby('neighbourhood_group_cleansed')['price'] \
    .mean() \
    .sort_values(ascending=False)


df_avg_nightly_by_neighborhood = pd.DataFrame(df_avg_nightly_by_neighborhood).sort_values(by=['neighbourhood_group_cleansed']).reset_index()
df_avg_nightly_by_neighborhood = df_avg_nightly_by_neighborhood.rename(columns={'neighbourhood_group_cleansed': 'neighborhood', 'price': 'avg nightly rate'})
df_avg_nightly_by_neighborhood
```

```python
#What is the average price by month?
df_avg_monthly = df_listings_small_calendar[df_listings_small_calendar['available'] == True] \
    .groupby('month')['price'] \
    .mean() \
    .sort_values(ascending=False)

#df_avg_monthly
df_avg_monthly = pd.DataFrame(df_avg_monthly).sort_values(by=['month']).reset_index()
df_avg_monthly = df_avg_monthly.rename(columns={'month': 'Month', 'price': 'Avg Rate'})
df_avg_monthly
```

```python
#What is the distribution of rental prices by neighborhood?
df_dist_prices_neighborhood = df_listings_small_calendar[df_listings_small_calendar['available'] == True] \
    .groupby(['neighbourhood_group_cleansed', 'price_quartile'])['date'] \
    .count() \
    .sort_values(ascending=False) \
    .rename(columns = {'neighbourhood_group_cleansed': 'neighborhood', 'date': 'count'})
```

```python
#What is the # of properties per host & annual value?
df_listings_by_host = df_listings_small_calendar.groupby('host_id') \
    .agg({'listing_id':'nunique', 'price': 'sum'}) \
    .reset_index() \
    .rename(columns={'host_id': 'Host ID', 'listing_id':'# of Locations', 'price': 'Annual Listing Value'})

df_listings_by_host = pd.DataFrame(df_listings_by_host).sort_values(by=['Annual Listing Value'], ascending=False)
df_listings_by_host
```

## Create Charts

### What is the average nightly rate by neighborhood?

```python
# Create plot size
f, ax = plt.subplots(figsize=(8,5))
# Set background style
sns.set_style('darkgrid')

color="coral"

# Plot data 
sns.barplot('neighborhood','avg nightly rate',data=df_avg_nightly_by_neighborhood,color=color,saturation=.5,)

# Add titles and labels
plt.title('Avg Nightly Rate by Neighborhood',fontsize=16)
plt.xlabel('Neighborhood',fontsize=14)
plt.ylabel('Nightly Rate',fontsize=14)

# Set x-axis labels and rotate vertical
ax.set_xticklabels(df_avg_nightly_by_neighborhood['neighborhood'], rotation = 90,fontsize=12)

# Move title and plot slightly further apart
ttl = ax.title
ttl.set_position([.5, 1.05])
plt.show()
```

```python
# Create plot size
f, ax = plt.subplots(figsize=(8,5))
# Set background style
sns.set_style('darkgrid')

color="coral"

# Plot data 
sns.barplot('Month','Avg Rate',data=df_avg_monthly,color=color,saturation=.5,)

# Add titles and labels
plt.title('Avg Nightly Rate by Month',fontsize=16)
plt.xlabel('Month',fontsize=14)
plt.ylabel('Avg Rate',fontsize=14)

# Set x-axis labels and rotate vertical
#ax.set_xticklabels(df_avg_nightly_by_neighborhood['neighborhood'], rotation = 90,fontsize=12)

# Move title and plot slightly further apart
ttl = ax.title
ttl.set_position([.5, 1.05])
plt.show()
```
